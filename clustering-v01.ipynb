L'approche proposée par Laio [1] permet de regrouper des milliards de données de manière simple et rapide. 

Le partitionnement de données est largement utilisé par exemple dans les études de marché particulièrement lorsque l'on travaille avec des données hétérogènes. Les études de marché utilisent le clustering pour partitionner la population globale des consommateurs en segments de marché et également pour mieux comprendre les relations entre les différents groupes de consommateurs. Le clustering sert également pour la segmentation du marché ou le développement de nouveaux produits par exemple.
 
## 1. Méthode de classification basée sur la distance
Les méthodes telles que [k-means](http://fr.wikipedia.org/wiki/Algorithme_des_k-moyennes), [k-medoids](http://fr.wikipedia.org/wiki/K-m%C3%A9do%C3%AFdes) [2][3] se basent sur le calcul de distance.
L'algorithme k-means partitionne un ensemble de données en k sous-groupes qui minimise la variance en minimisant la distance entre les éléments d'un même cluster.

Considerons l'exemple tiré du livre de Webb [4]. Soit six points numérotés de 1 à 6 dans $\mathbb{R}^2$, on pose $k=2$, donc nous obtiendrons au final deux classes. 


1. Prenons $5$ et $6$ comme centre de gravité initiale de chaque sous-groupe.
2. On regroupe les différents points aux deux classes: Les points les plus proches de $5$ (au sens de la distance euclidienne) que de $6$ seront dans le premier groupe, les autres dans le second. On obtient un premier groupe constitué de $(1, 2, 3, 4, 5)$ et un second de $(6)$ seulement.

3. On Calcule le centre de gravité des deux classes: On obtient $A_1$ qui a pour coordonnées $(1.6, 2.4)$ et le second $B1$ qui est le point $6$, dont les coordonnées sont $(5, 2)$.

4. On réitère la phase 1 et 2 : les six points sont redistribués en fonction des centres de gravité $A_1$ et $B_1$. On obtient les groupes $(1, 2, 3, 4)$ et $(5, 6)$ avec comme centre de gravité $A_2$ et $B_2$ respectivement.

Une nouvelle itération ne modifie plus la partition, l'algorithme a convergé.



La méthode [k-medoides](http://fr.wikipedia.org/wiki/K-m%C3%A9do%C3%AFdes) cherche à minimiser la distance entre les points d'un cluster et le ménoïde (le point le plus proche du centre du cluster). 

Les algorithmes k-means et k-medoides ne sont pas capable de détecter des clusters non sphériques. L'un des défauts majeur de ces techniques, est le fait de devoir défnir le nombre de clusters à l'avance en fixant $k$.

## 2. Méthode de classification basée sur la densité

DBSCAN et MeanèShift font partie des algorithmes les plus connus. Ces méthodes sont capable de découvrir des clusters non sphériques. Mean-Shift par exemple consiste à :

1. Définir un point de départ P comme centre du cluster.

2. Regrouper l'ensemble des points qui se trouvent au voisinage de P et former ainsi un sous-groupe de points.

3. Calculer M, le max de la fonction de densité du sous-groupe. 

4. Deplacer P (le centre du cluster) vers M. qui sera ainsi le nouveau centre de notre clsuter.

5. Réitérer les étapes 2, 3, et 4 jusqu'à convergence. P finira par converger vers le maximum local de la fonction de densité. [Densité]




#### Ci-dessous un exemple du clustering avec mean-shift :

```python
import numpy as np

from sklearn.cluster import MeanShift, estimate_bandwidth
from sklearn.datasets.samples_generator import make_blobs

###############################################################################
# Géneration de 20000 points au tour de "centers"
centers = [[1, 1], [-1, -1], [1, -1]] 
X, _ = make_blobs(n_samples=20000, centers=centers, cluster_std=0.6)

###############################################################################
# Clustering avec MeanShift

# Bande passante à passé a la méthode Mean-Shift
largeur_bande = estimate_bandwidth(X, quantile=0.2, n_samples=500)

# Initiation de la méthode Mean-Shift
ms = MeanShift(bandwidth=largeur_bande, bin_seeding=True)
# Execution du Clustering
ms.fit(X)
labels = ms.labels_
centres = ms.cluster_centers_

labels_unique = np.unique(labels)
nombres_clusters= len(labels_unique)

print("Nombre de Clusters estimés : %d" % nombres_clusters)

###############################################################################
# Visualisation des resultats
%matplotlib inline
import matplotlib.pyplot as plt
from itertools import cycle

plt.figure(1)
plt.clf()

colors = cycle('bgrcmykbgrcmykbgrcmykbgrcmyk')
for k, col in zip(range(nombres_clusters), colors):
    my_members = labels == k
    centre = centres[k]
    plt.plot(X[my_members, 0], X[my_members, 1], col + '.')
    plt.plot(centre[0], centre[1], 'o', markerfacecolor=col,
             markeredgecolor='k', markersize=14)

plt.show()
```

## 3. La méthode de l'article : Clustering by fast search and find of density peaks

La méthode proposée par A. Laio & A. Rodriguez [5] est similaire à celles des k-medoids, elle se base uniquement sur la distance entre les points. Comme les méthodes DBSCAN et Mean-shift, cette méthode est capable de détecter les clusters non-sphériques et de déterminer le nombre de clusters de manière automatique. 

Cette méthode est basée sur l'idée que les centres des clusters ont une densité locale plus élevée que celle de leur voisins les plus proches (density peak). Pour cela, il faut calculer pour chaque point $i$ sa densité locale $\rho_i$ et la distance $\delta_i$ qui le sépare des points à plus forte densité locale.

$$\rho_i=\sum_{j}\chi(d_{ij}-d_{c}) $$


$d_{ij} =$  la distance entre les points $i$ et $j$.

$d_{c}$ : Le seuil de voisinage, il faut définir un seuil $d_{c}$ de telle sorte que la moyenne des voisins soit entre 1 et  2% sur la totalité des points.

$\chi(x) = 1$ si $x<0$ sinon $\chi(x)=0$


Si un point $j$ se trouve à une distance $d_{ij}$ de $i$ inférieur à $d_{c}$, alors $\chi(d_{ij}-d{c})$ = 1. C'est pourquoi la densité $\rho_{i}$ est égale au nombre de points se trouvant à une distance de $i$ inférieur à $d_{c}$. Un avantage important sur l'algorithme Mean-Shift qui d'abord projete les données sur un espace vectoriel pour chercher la densité maximale ([de la fonction de densité](http://fr.wikipedia.org/wiki/Densit%C3%A9_de_probabilit%C3%A9)) et cela pour chaque point. Or les calculs vectoriels sont très gourmands en temps de calcul.
    


$\delta_i$ : La distance minimale entre i et tout autre point ayant une densité supérieur à celle de $i$.

$$\delta_i = min_{j:\rho_j>\rho_i}(d_{ij})$$

$\delta_i = max_j(d_ij)$ pour les ponints à très forte densité. 

Ainsi, les centres des clusters seront reconnus grâce à une forte densité mais aussi, une valeur de $\delta_{i}$ très large, plus large que $d_c$. Le centre du cluster est forcement entouré de points à densité inférieur.
Après avoir découvert les centres des clusters, chaque point restant est assigné au centre le plus proche. L'affectation d'un  point à un cluster donné se fait en une seule étape, contrairement aux méthodes basées sur la convergence comme nous l'avons vu avec l'algorithme Mean-shift.

Les points 1 et 10 ont une densité élevée et un $\delta$ très large. C'est pourquoi ces deux points sont réciproquement les centres des clusters bleu et rouge. Les points 9 et 10 ont la même densité $\rho$ mais des $\delta$ différents. On constate que le point 9 a un $\delta$ assez faible, puisqu'il est entouré de voisin à plus forte densité.Il appartient donc au cluster bleu. Les points 26, 27 et 28 ont un $\delta$ assez large mais une densité faible. Ce sont des centres de clusters à un seul élément.

Cette nouvelle approche basée uniquement sur la distance et proposant une affectation de chaque point en une seule phase, permet de catégoriser des pếta-octets de données rapidement et efficacement.
